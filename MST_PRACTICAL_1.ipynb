frequency of characters in a string

```python
def word_frequency(string):

    string = string.lower().replace(',', '').replace('.', '').replace('!', '').replace('?', '')

    words = string.split()

    frequency = {}


    for word in words:

        frequency[word] = frequency.get(word, 0) + 1

    return frequency

input_string = string
result = word_frequency(input_string)

for word, freq in result.items():
    print(f"Word '{word}' occurs {freq} times.")
```

main words in the string

```python


def find_main_words(text):

    words = word_tokenize(text)

    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))

    filtered_words = [word for word in words if word.lower() not in stop_words]

    return filtered_words

text = input_string

main_words = find_main_words(text)

print("Main words in the text:")
print(main_words)
```

```python
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
nltk.download('averaged_perceptron_tagger')
```

```python
train_text=(string)
sample_text=(string)
```

```python
Customsent_tokenizer=PunktSentenceTokenizer(train_text)
tokenized=Customsent_tokenizer.tokenize(sample_text)
```

```python
def process_content():
    try:
        for i in tokenized[:5]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)

            chunkGram = r"""Chunk: {<RB.?>*<VB.?>*<NNP>+<NN>?}"""
            chunkParser = nltk.RegexpParser(chunkGram)
            chunked = chunkParser.parse(tagged)
            print(chunked)

    except Exception as e:
        print(str(e))
```

```python
process_content()
```
